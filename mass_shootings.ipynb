## Analysis per year

```python
def df_year(year):
  """Data Frame with only data about the year selected"""
  condition = mass_shootings_usa["Date"].str.contains(year)
  data_frame = mass_shootings_usa[condition]
  return data_frame
```

```python
def more_killed(df, year):
  """Mass shooting bloodiest in that year"""
  df_bloodiest = df[df["Murders"] == np.max(df["Murders"], axis=0)]
  df_bloodiest = df_bloodiest.loc[:, ["State", "City", "Murders"]]
  df_bloodiest["Year"] = year
  return df_bloodiest
```

```python
def more_injured(df, year):
  """Mass shooting with more injured"""
  df_injured = df[df["Injured"] == np.max(df["Injured"], axis=0)]
  df_injured = df_injured.iloc[:, [2, 3, 6]]
  df_injured["Year"] = year
  return df_injured
```

```python
def total_victims(df, year):
  df_victims = df.loc[:, ["State", "City", "Murders", "Injured"]]
  df_victims["Total_victims"] = df_victims["Murders"] + df_victims["Injured"]
  df_victims["Year"] = year
  df_victims = df_victims[df_victims["Total_victims"] == np.max(df_victims["Total_victims"], axis=0)]
  return df_victims
```

```python
def city_mass_shooting(df):
  """City with more mass shooting"""
  city = df["City"].value_counts()
  return city
```

```python
def states_mass_shooting(df):
  """In how many states had mass shootings."""
  q_states = df["State"].value_counts()
  return q_states.shape[0]
```

```python
def months_mass_shootings(df):
  """In which month had more mass shootings"""
  months_list = [
          "January", "February", "March", "April", 
          "May", "June", "July", "August", 
          "September", "October", "Novemeber", "December"
]
  frequency = {}
  for m in months_list:
    df_month = df[df['Date'].str.contains(m)]
    frequency[m] = df_month.shape[0]
  max_frequency = max(frequency.values())
  months = []
  for k, v in frequency.items():
    if v == max_frequency:
      months.append(k)
  return months, max_frequency
```
